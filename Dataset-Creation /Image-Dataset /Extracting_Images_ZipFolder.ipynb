```python
import os
import zipfile

source_folder='/content/drive/MyDrive/2023'
destination_folder='/content/drive/MyDrive/Colab Notebooks'
# Uncomment the line below if you to check the existence of the destination folder
os.makedirs(destination_folder, exist_ok=True)

# Uncomment the line below if you are not sure about the existence of the destination folder
# os.makedirs(destination_folder, exist_ok=True)

# File names to keep as ceating dataset with these type of images for better segmentation
filenames_to_keep = '12_UE.jpg'

# Iterate through the ZIP files in the source folder
for filename in os.listdir(source_folder):
    if filename.endswith('.zip'):
        zip_filepath = os.path.join(source_folder, filename)
        try:
            with zipfile.ZipFile(zip_filepath, 'r') as zip_ref:
                for file_info in zip_ref.infolist():
                   #print(f"Checking file: {file_info.filename}")
                   #print(f"Filenames to keep: {filenames_to_keep}")
                   if filenames_to_keep in file_info.filename :
                        #print(f"Found desired file in ZIP: {file_info.filename}")
                        zip_ref.extract(file_info, destination_folder)
                        print(f"Extracted {file_info.filename} to {destination_folder}")
                        extracted_file_path = os.path.join(destination_folder, os.path.basename(file_info.filename))
                        if os.path.exists(extracted_file_path):
                            print(f"File successfully saved at {extracted_file_path}")
                        else:
                            print(f"Failed to save file at {extracted_file_path}")
            print(f"Extraction completed for {zip_filepath}")

        except zipfile.BadZipFile:
            print(f"The file at {zip_filepath} is not a valid zip file or is corrupted.")
        except OSError as e:
            print(f"OS error occurred: {e}")
        except Exception as e:
            print(f"An unexpected error occurred: {e}")
```
